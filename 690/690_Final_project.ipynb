<h3><center>What is Renewable Energy?</center></h3>
<div style="text-align: justify">First, we need to know what renewable energy is. It is energy from sources that are naturally recharge but its flow is limited. Renewable energy is energy derived from natural resources that refilled themselves in less than a human lifetime without depleting the planet’s resources. These resources such as sunlight, wind, rain, tides, waves, biomass, and thermal energy stored in the earth’s crust have the benefit of being available in one form or another nearly everywhere. They are virtually inexhaustible. And what is even more important, they cause little climate or environmental damage.</div>
<h3><center>Why Renewable Energy is Important?</center></h3>
<div style="text-align: justify">Renewable energy provides reliable power supplies and fuel diversification, which enhance energy security, lower risk of fuel spills, and reduce the need for imported fuels. Renewable energy also helps conserve natural resources.</div>

<h3><center>Electricity Production from Renewable Sources</center></h3>
<div style="text-align: justify">We first consider electricity production in kilo watt hour (kWh) from renewable sources. This production excludes hydroelectric power. It considers sources including geothermal, solar, tides, wind, biomass, and biofuels.</div>

<div style="text-align: justify">Figure 1 compares the produced electricity energy against the population of each country. As one can see India produces less renewable energy considering its population. On the other hand, the United States produces the highest amount of energy considering its population. The amount of produced renewable energy in China is proportionate to its population. We will take the following steps to create Figure 1.</div>

```python
# Importing necessary packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Reading data of electricity production from renewable sources(KWH)
re_kwh = pd.read_csv('./datasets-690-project/new_datasets/Electricity production from renewable sources, excluding hydroelectric (kWh).csv')
re_kwh.head(3)
```

```python
# Reading data of population
countries_popul = pd.read_csv('./datasets-690-project/new_datasets/population.csv')
countries_popul.head(3)
```

```python
# Extracting only the year 2015
re_kwh_year_2015 = re_kwh[re_kwh['Year']== 2015]
```

```python
# Merging population and renewable energy tables
re_kwh_popul = pd.merge(re_kwh_year_2015, countries_popul, on='Country Name', how='left')
re_kwh_popul.head(3)
```

```python
# Normalize population based on the highest population
# These percentages are used for changig the buble size of the scatter plot
pop = re_kwh_popul['2015 [YR2015]']
pop_pct = pop/max(pop)
```

```python
# Scatter plot of the population vs the produced renewable energy
countries = re_kwh_popul['Country Name'].unique().tolist()
fig = plt.figure(figsize = (12,8))
legend = []
for i, country in enumerate(countries):
    country_data = re_kwh_popul[re_kwh_popul['Country Name']== country]
    x = country_data['value']
    y = country_data['2015 [YR2015]']
    plt.scatter(x, y, s = 1500*pop_pct[i])
    plt.text(x, y, f"{country}", fontsize = 15)
    legend.append(country)
plt.title('Figure 1: Renewable Energy Production Versus Population (2015)', pad=20, fontsize= 20)
plt.xlabel('Renewable Energy Production(kwh)', fontsize= 15)
plt.ylabel('Population 2015', fontsize= 15)
plt.legend(legend,  bbox_to_anchor=(1 ,1), title = "Countries", prop = {'size':20})
plt.show()
```

<div style="text-align: justify">As you can see in Figure 2, the United States and China produce the most renewable energy among all powerful economies. The question is whether these productions are a considerable amount of the total productions of electricity for each country. To answer this question, we will look at the percentage of renewable energy production from the total energy production in Figure 3. We will take the following steps to create Figure 2.</div>

```python
# Extracting rows including renewable energy production indicator
re_kwh_year_2015 = re_kwh_year_2015[re_kwh_year_2015['indicator']== 'EG.ELC.RNWX.KH']
re_kwh_year_2015.head(3)
```

```python
# Plotting pie chart of renewable energy production
labels = re_kwh_year_2015['Country Name']
ax = re_kwh_year_2015.plot.pie(y='value', labels = labels, autopct='%1.1f%%' , figsize=(8, 8))
ax.set_ylabel('')
plt.title('Figure 2: 2015 Renewable energy production (kWh)', fontsize = 20)
fig = plt.gcf()
# Relocating the legend box
plt.legend(title = "Countries", bbox_to_anchor=(1 ,1))
plt.show()
```

<div style="text-align: justify">As you can see in Figure 3 the proportion of production of the renewable energy in the United State and China is much smaller that their total electricity production. For example, about 8% of the total energy produced in the United States is renewable energy, while 25% of produced total energy in Germany is renewable energy. Another observation is that the proportion of the produced renewable energy in European countries is higher than other regions.</div>

```python
# Reading data of the proportion of production of the renewable energy
re_percent = pd.read_csv('./datasets-690-project/new_datasets/Electricity production from renewable sources, excluding hydroelectric (% of total).csv')
re_percent.head(3)
```

```python
# Extracting only the year 2015
re_percent_year_2015 = re_percent[re_percent['Year']== 2015]
```

```python
# Plotting the bar chart of proportion of production of the renewable energy
ax = re_percent_year_2015.sort_values('value', ascending=False).plot.bar(x='Country Name',
                                                                         y='value',
                                                                         figsize=(12,4),
                                                                         width = 0.8,
                                                                         legend=False)
ax.set_xlabel("Countries", fontsize= 12)
ax.set_ylabel("Renewable energy production percentage", fontsize= 10)
ax.set_title("Figure 3: 2015 Renewable energy production (% of total)", fontsize= 20)
x = re_percent_year_2015.sort_values('value', ascending=False)['Country Name']
ax.set_xticklabels(x, rotation = 30)
plt.show()
```

<h3><center>Which Country has the Fastest Pace in Producing Renewable Energy Historically?</center></h3>

As Figure 4 shows China has ramped up its renewable energy production after 2012. However, other countries have increased their production with a constant rate.


```python
# Plotting historical trend in producing renewable energy (2010-2015)
countries = re_kwh['Country Name'].unique().tolist()
fig = plt.figure(figsize = (12,6))
legend = []
for country in countries:
    country_data = re_kwh[re_kwh['Country Name']== country]
    x = country_data['Year']
    y = country_data['value']
    plt.plot(x, y)
    legend.append(country)
plt.title('Figure 4: Historical trend in producing renewable energy (2010-2015)',
          pad=20, fontsize= 20)
plt.xlabel('Year', fontsize= 15)
plt.ylabel('Electricity production from renewable sources(kwh)', fontsize= 12)
plt.legend(legend,  bbox_to_anchor=(1 ,1), title = "Countries")
plt.show()
```

<h3><center>In Which Country the Renewable Energy Consumption Proportion is the Highest?</center></h3>
It is true that producing renewable energy is very important but using it is important as well. As Figure 5 indicates India is best country in utilizing renewable energy among all nine countries. About 35% of its total consumption is renewable energy.


```python
# Reading data of the reneable enery consumption
re_consumption = pd.read_csv('./datasets-690-project/new_datasets/Reneable enery consumption.csv')
re_consumption.head(3)
```

```python
# Extracting only the year 2015
re_consumption_year_2015 = re_consumption[re_consumption['Year']== 2015]
```

```python
# Plotting share of renewable energy consumption from total energy
ax = re_consumption_year_2015.sort_values('value', ascending=False).plot.bar(x='Country Name',
                                                                             y='value',
                                                                             figsize=(12,4),
                                                                             width = 0.8,
                                                                            legend=False)
ax.set_xlabel("Countries", fontsize= 12)
ax.set_ylabel("Renewable energy consumption(%)", fontsize= 12)
ax.set_title("Figure 5: Share of renewable energy consumption from total energy (2015)",
             fontsize= 20)
x = re_consumption_year_2015.sort_values('value', ascending=False)['Country Name']
ax.set_xticklabels(x, rotation = 30)
plt.show()
```
